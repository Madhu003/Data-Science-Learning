### Question 2.

Scatter  graph is a way of showing data visually. where x & y axis data are mentioned as single point. Which are multiple in graph as there may be lots of details accordingly dots will be mentioned. And that is called scattred plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.xlabel("Data for x axis")
plt.ylabel("Data for y axis")
plt.title("This is my first plot")
```
