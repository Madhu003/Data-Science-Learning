### Question 2.

Scatter  graph is a way of showing data visually. where x & y axis data are mentioned as single point. Which are multiple in graph as there may be lots of details accordingly dots will be mentioned. And that is called scattred plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.xlabel("Data for x axis")
plt.ylabel("Data for y axis")
plt.title("This is my first plot")
```

### Question 3.

Subplot is a feature from matplotlib which is used to create multiple plot in single place.

```python
# fig, axes = plt.subplots(nrows=1, ncols=2)

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x1, y1)

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x2, y2)

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x3, y3)

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x4, y4)

# Show the figure
plt.show()
```

### Question 4.

Bar plot is a way to plot graph where data is visualize like bar in which y axis value define height of bar and x axis for positioning the bar. That is basically that kind of graph is used to show category based data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Companies")
plt.title("Horizontal bar graph")
plt.show()
```

### Question 5.

As I understood by reading from internet box graph is a graph in which data is used after calculation in median, min & max.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# print(box1)
# print(box2)

fig, ax = plt.subplots()
ax.boxplot(box1)
ax.boxplot(box2)

ax.set_title('Box plot of random data')
ax.set_xlabel('box1')
ax.set_ylabel('box2')

plt.show()
```
