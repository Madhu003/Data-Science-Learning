### Question 5.

1. **Range** - Range just a way to provide how much data is spreaded. Here need to substract minumum from maximum.

```python
max = np3.max()
min = np3.min()

print("Range => ", max - min)
```

2. **variance** - Its a sum of sequared difference b/w number & mean.

```python
np3.var()
```

3. **standard deviation** - That is just sequare root of varience. It's a measure of extent to which data varies from mean. 

```python
np3.std()
```
