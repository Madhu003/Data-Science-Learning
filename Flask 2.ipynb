**GET** & **POST** both are methods to communicate between server & client. 

One is for getting data from server with specific parameter for that we use **GET** (Ex. google search). And in that method execution we can see data in form of query parameter.

**POST** is use to send data in body like we have filled user information in form and wanted to send that to server we can use POST for that and in that action data will be send in body so it won't be visible and we can say that this is a secure way.

### Question 2.

**request** is a function which is use to access properties related to API like parameter & data sent by user in API.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        
        if ops == 'add':
            return num1 + num2
        if ops == 'subtract':
            return num1 - num2
        if ops == 'multiply':
            return num1 * num2
        if ops == 'divide':
            return num1 / num2
```

### Question 4.

**render_template()** is function is to take a html file and render to on web app as required with different params.

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    name = 'John'
    age = 30
    return render_template('home.html', name=name, age=age)

# here html file we be something like with placeholders of given parameter 
# which will replaced with passed value at run time

"""
home.html

<!DOCTYPE html>
<html>
<head>
    <title>Home Page</title>
</head>
<body>
    <h1>Welcome, {{ name }}!</h1>
    <p>You are {{ age }} years old.</p>
</body>
</html>
"""
```
