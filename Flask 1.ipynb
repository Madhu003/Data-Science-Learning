**url_for()** is use for URL binding and by that url will be generated as per function name.
ANd function argument is used as api params.

```python
@app.route('/user/<username>')
def user_profile(username):
    return f'Profile page of {username}'

with app.test_request_context():
    print(url_for('user_profile', username='john'))  # Output: '/user/john' 
```

that will work only in with block and we can use that for API testing.
